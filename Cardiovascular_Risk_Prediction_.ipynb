````python
# storing the column that contains null values
null_column_list= ['glucose','education','BPMeds','totChol','cigsPerDay','BMI','heartRate']
# plotting box plot
plt.figure(figsize=(10,8))
df[null_column_list].boxplot()
````

````python
# Iterate over the null column list and plot each column's distribution
colors = sns.color_palette("rocket", len(null_column_list))


fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10, 8))


axes = axes.flatten()


for i, column in enumerate(null_column_list):
    ax = axes[i]

    sns.distplot(df[column], ax=ax, color=colors[i])
    ax.set_title(column)

for j in range(len(null_column_list), len(axes)):
    axes[j].remove()

plt.show()
````

````python
# Imputing missing values with median or mode
df.fillna({'glucose': df['glucose'].median(),
           'education': df['education'].mode()[0],
           'BPMeds': df['BPMeds'].mode()[0],
           'totChol': df['totChol'].median(),
           'cigsPerDay': df['cigsPerDay'].median(),
           'BMI': df['BMI'].median(),
           'heartRate': df['heartRate'].median()}, inplace=True)
````

### What all manipulations have you done and insights you found?

We had some information missing in our data, like numbers for glucose, cholesterol, smoking, weight, and heart rate, and also for things like education and whether someone takes certain medications. To fill in these gaps, we used two methods.

For the numbers like glucose, cholesterol, smoking amount, weight, and heart rate, we looked at the ones we already had and found the middle value (not too big or too small). We used that middle value to fill in the missing spots.

For things like education and medication info, we looked at what was already there and figured out what value appeared the most. Then, we used that popular value to complete the missing information.

These methods are common because they work well for different types of missing info. Using the middle value (median) helps when there are unusual numbers, and using the most popular value (mode) works when things are in categories or have only a few choices.
